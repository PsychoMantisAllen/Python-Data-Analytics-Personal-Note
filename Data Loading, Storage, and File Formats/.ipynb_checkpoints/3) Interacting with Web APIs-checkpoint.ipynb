One recommended method to access public APIs is the [requests package](https://requests.readthedocs.io/en/master/).

Example: to find the last 30 Github issues for pandas on Github, we can make a `GET` HTTP request using the add-on `requests` library: 

```python
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
```

```python
resp
```

```python
data = resp.json()
#return a dictionary containing JSON parsed into native Python objects
```

```python
data[0]['title']
```

All data found in Github issue page is stored except for the *comments*. We can pass it to DataFrame and extract our fields of interest:

```python
import pandas as pd
issues = pd.DataFrame(data, columns = ['number', 'title', 
                                       'labels', 'state'])
issues
```
